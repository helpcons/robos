```python
import win32com.client

# Inicia uma instância do Outlook
outlook = win32com.client.Dispatch("Outlook.Application")
namespace = outlook.GetNamespace("MAPI")

# Acessa a pasta de Caixa de Entrada
inbox = namespace.GetDefaultFolder(6)  # 6 é o código para a Caixa de Entrada

# Nome da subpasta que você quer acessar
subfolder_name = "Projetos"

try:
    # Acessa a subpasta
    subfolder = inbox.Folders[subfolder_name]

    # Lê os emails na subpasta
    messages = subfolder.Items
    for message in messages:
        print("Assunto:", message.Subject)
        print("Remetente:", message.SenderName)
        print("Data:", message.ReceivedTime)
        print("Corpo:", message.Body)
        print("----------")
except Exception as e:
    print(f"Erro ao acessar a subpasta '{subfolder_name}': {str(e)}")
```

```python
import win32com.client
import re
import pandas as pd

# Função para extrair URLs do corpo do email
def extract_urls(text):
    url_pattern = r'(https?://[^\s]+)'
    urls = re.findall(url_pattern, text)
    return urls

# Inicia uma instância do Outlook
outlook = win32com.client.Dispatch("Outlook.Application")
namespace = outlook.GetNamespace("MAPI")

# Nome da conta que você quer acessar
account_name = "lopesdev01@gmail.com"

# Encontra a conta específica
account = None
for acc in namespace.Folders:
    if acc.Name == account_name:
        account = acc
        break

# Lista para armazenar os dados
data = []

if account:
    # Acessa a Caixa de Entrada dessa conta
    inbox = account.Folders["Caixa de Entrada"]
    
    # Lê os emails na Caixa de Entrada
    messages = inbox.Items
    for message in messages:
        subject = message.Subject
        received_time = str(message.ReceivedTime)
        if " | Landing Page" in subject or "Landing | Lançamento" in subject:
            body = message.Body
            urls = extract_urls(body)
            for url in urls:
                data.append([received_time, subject, url])
else:
    print(f"A conta '{account_name}' não foi encontrada.")

# Cria um DataFrame com os dados
df = pd.DataFrame(data, columns=["Data de Recebimento", "Assunto", "URL"])

# Exibe o DataFrame
print(df)

# Salva o DataFrame em um arquivo CSV
df.to_excel("emails_urls.xlsx", index=False)
```
