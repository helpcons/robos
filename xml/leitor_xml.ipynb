```python
import requests
import xml.etree.ElementTree as ET

# Baixar o arquivo XML da URL
url = 'https://cdn.lopesrj.com.br/ext/site.xml'
response = requests.get(url)

# Verifica se o arquivo foi baixado com sucesso
if response.status_code == 200:
    xml_content = response.content
    
    # Parsear o conteúdo XML
    tree = ET.ElementTree(ET.fromstring(xml_content))
    root = tree.getroot()
    
    # Exemplo de como percorrer os elementos
    for elem in root.findall('.//Imovel'):  # Supondo que os dados estejam em <item>
        titulo = elem.find('CodigoImovel').text
        preco = elem.find('TipoImovel').text
        print(f'Título: {titulo}, Preço: {preco}')
else:
    print(f'Erro ao baixar o arquivo XML. Status code: {response.status_code}')
```

```python
import requests
import xml.etree.ElementTree as ET

# Baixar o arquivo XML da URL
url = 'https://cdn.lopesrj.com.br/ext/site.xml'
response = requests.get(url)

# Verifica se o arquivo foi baixado com sucesso
if response.status_code == 200:
    xml_content = response.content
    
    # Parsear o conteúdo XML
    tree = ET.ElementTree(ET.fromstring(xml_content))
    root = tree.getroot()
    
    # Busca pelo código específico
    codigo_busca = 'BC310820'
    encontrado = False
    
    # Usar enumerate para saber a posição (índice) de cada imóvel
    for idx, imovel in enumerate(root.findall('.//Imovel')):
        codigo_imovel = imovel.find('CodigoImovel').text
        if codigo_imovel == codigo_busca:
            encontrado = True
            print(f'Imóvel encontrado na posição {idx} com o código: {codigo_imovel}')
            # Exibir outras informações, se necessário
            endereco = imovel.find('Endereco').text
            cidade = imovel.find('Cidade').text
            preco_venda = imovel.find('PrecoVenda').text
            print(f'Endereço: {endereco}, Cidade: {cidade}, Preço de Venda: {preco_venda}')
            break
    
    if not encontrado:
        print(f'Código do imóvel {codigo_busca} não foi encontrado.')
else:
    print(f'Erro ao baixar o arquivo XML. Status code: {response.status_code}')
```
